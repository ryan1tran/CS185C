```python
import os
import re
```

```python
path = "C:\\Users\\Michael\\Desktop\\CS 185C\\Malware\\samples"

files = []

#root, directory, folder
for r, d, f in os.walk(path):
    for file in f:
        if file.endswith('.txt'):#get all txt file names
            files.append(os.path.join(r, file))
```

```python
#deprecated
def count_file(file_name):
    freq = {}
    file = open(file_name, 'r')
    for line in file.readlines():
        str = re.sub('[^a-z]', '', line.lower())#remove all non alpha characters
        if str in freq.keys():#increment frequency count
            freq[str] = freq[str] + 1
        else:#if opcode not seen yet, add it
            freq[str] = 1
    file.close()
    return freq
```

```python
def tuple_to_string(opcode_tuple):
    string = ""
    for tup in opcode_tuple:
        string += tup
    return string
```

```python
def count_file_tuples(file_name, num):
    freq = {}
    opcode_tuple = []
    file = open(file_name, 'r')
    lines = file.readlines()
    for i in range(num):
        line = re.sub('[^a-z]', '', lines[i].lower())#remove all non alpha characters
        opcode_tuple.append(line)
    for i in range(num, len(lines)):
        string = tuple_to_string(opcode_tuple)
        if string in freq.keys(): #increment frequency count
            freq[string] = freq[string] + 1
        else: #if opcode not seen yet, add it
            freq[string] = 1
        
        opcode_tuple.pop(0)
        line = re.sub('[^a-z]', '', lines[i].lower())#remove all non alpha characters
        opcode_tuple.append(line)
    file.close
    return freq
```

```python
def merg_dict(dict1, dict2):
    dict3 = {**dict1, **dict2}
    for key, value in dict3.items():
        if key in dict1 and key in dict2:
            dict3[key] = dict1[key] + dict2[key]
    return dict3
```

```python
#deprecated
def count_files(files):
    freqs = {}
    for file_name in files:#merge all frequency across malware families
        freqs = merg_dict(freqs, count_file(file_name))
    return freqs
```

```python
def count_files_tuples(files, num):
    freqs = {}
    for file_name in files:#merge all frequency across malware families
        freqs = merg_dict(freqs, count_file_tuples(file_name, num))
    return freqs
```

```python
num_tuple = 6
freq = count_files_tuples(files, num_tuple)
#for key in freq.keys():
#    print(f"{key}: {freq[key]}")
```

```python
sorted_freq = dict(sorted(freq.items(), key=lambda x: x[1], reverse=True))
```

```python
for key in sorted_freq.keys():
    if(freq[key] < 30):
        break
    print(f"{key}: {freq[key]}")
```

```python
freq_file = open("C:\\Users\\Michael\\Desktop\\CS 185C\\Malware\\opcode Frequencies {}-tuple.txt".format(num_tuple), 'w')
for key in sorted_freq.keys():#save frequency counts to file
    if(freq[key] < 30):
        break
    freq_file.write(f"{key}: {freq[key]}\n")
freq_file.close()
```
